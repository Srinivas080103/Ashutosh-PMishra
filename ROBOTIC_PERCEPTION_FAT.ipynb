```python
#SIFT
import cv2
import matplotlib.pyplot as plt

# Load the image
img_path = 'sample.png'  # Replace with your image file path
img = cv2.imread(img_path)

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Initialize the SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints, descriptors = sift.detectAndCompute(gray, None)

# Draw keypoints on the image
img_with_keypoints = cv2.drawKeypoints(img, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the original image and the image with keypoints
fig, ax1 = plt.subplots(1, 2, figsize=(12, 6))
ax1[0].axis('off')
ax1[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1[0].set_title('Original Image')
ax1[1].axis('off')
ax1[1].imshow(cv2.cvtColor(img_with_keypoints, cv2.COLOR_BGR2RGB))
ax1[1].set_title('Image with SIFT Keypoints')
plt.show()
```

```python
#Contrast stretching of a low contrast image, Histogram, and Histogram Equalization.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('pics.png', 0)

# Check if the image is loaded correctly
if image is None:
    print("Error: Image not found or unable to load.")
else:
    # Contrast Stretching
    min_val = np.min(image)
    max_val = np.max(image)
    stretched_image = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)

    # Histogram Calculation
    def plot_histogram(image, title):
        plt.figure()
        plt.title(title)
        plt.xlabel("Intensity Value")
        plt.ylabel("Frequency")
        plt.hist(image.ravel(), bins=256, range=[0, 256])
        plt.show()

    # Histogram Equalization
    equalized_image = cv2.equalizeHist(image)

    # Display the results
    cv2.imshow("Original Image", image)
    cv2.imshow("Contrast Stretched Image", stretched_image)
    cv2.imshow("Histogram Equalized Image", equalized_image)

    # Plot histograms
    plot_histogram(image, "Original Image Histogram")
    plot_histogram(stretched_image, "Contrast Stretched Image Histogram")
    plot_histogram(equalized_image, "Histogram Equalized Image Histogram")

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#Implement sobel edge detection from scratch.
Code:

import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage import convolve

def sobel_edge_detection(image):
    # Convert to grayscale if the image is not already in grayscale
    if len(image.shape) == 3:
        image = np.dot(image[..., :3], [0.2989, 0.5870, 0.1140])

    # Define Sobel kernels
    sobel_x = np.array([[-1., 0., 1.],
                        [-2., 0., 2.],
                        [-1., 0., 1.]])
    sobel_y = np.array([[-1., -2., -1.],
                        [ 0.,  0.,  0.],
                        [ 1.,  2.,  1.]])

    # Apply convolution with Sobel kernels
    gradient_x = convolve(image, sobel_x)
    gradient_y = convolve(image, sobel_y)

    # Calculate gradient magnitude
    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
    gradient_magnitude = (gradient_magnitude / np.max(gradient_magnitude) * 255).astype(np.uint8)

    return gradient_magnitude

# Load the image
image = plt.imread('pics.png')

# Apply Sobel edge detection
edges = sobel_edge_detection(image)

# Display the results
plt.figure(figsize=(16, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('Sobel Edge Detection')
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

```python
### 1. **Clustering-based Segmentation (K-Means)**
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Read the image
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Reshape image to a 2D array of pixels and 3 color values (RGB)
pixel_values = image.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

# Define criteria and apply KMeans
k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixel_values)

# Convert the labels back to the original image shape
labels = kmeans.labels_
segmented_image = labels.reshape(image.shape[:2])

# Display the result
plt.imshow(segmented_image)
plt.show()
```

```python
### 2. **Mean Shift Segmentation**

import cv2
import numpy as np
from sklearn.cluster import MeanShift, estimate_bandwidth
import matplotlib.pyplot as plt

# Read and convert the image to RGB
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Reshape the image into 2D array
pixel_values = image.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

# Estimate the bandwidth (window size) for mean shift
bandwidth = estimate_bandwidth(pixel_values, quantile=0.2, n_samples=500)

# Perform MeanShift clustering
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(pixel_values)

# Get cluster labels and reshape to original image size
labels = ms.labels_
segmented_image = labels.reshape(image.shape[:2])

# Display the result
plt.imshow(segmented_image)
plt.show()
```

```python
### 3. **Graph-based Segmentation (Felzenszwalb's Algorithm)**

from skimage import io, segmentation, color
from skimage.future import graph
from skimage.segmentation import felzenszwalb
import matplotlib.pyplot as plt

# Read the image
image = io.imread('image.jpg')

# Apply Felzenszwalb's algorithm
segments_fz = felzenszwalb(image, scale=100, sigma=0.5, min_size=50)

# Visualize the segmented output
plt.imshow(color.label2rgb(segments_fz, image, kind='avg'))
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def add_noise(image, noise_factor=0.1):
noisy_image = image + noise_factor * np.random.randn(*image.shape)
return np.clip(noisy_image, 0, 1)

def median_filter(image, filter_size):
padding = filter_size // 2
padded_image = np.pad(image, ((padding, padding), (padding, padding)), mode='reflect')
filtered_image = np.zeros_like(image)
for i in range(image.shape[0]):
for j in range(image.shape[1]):
window = padded_image[i:i+filter_size, j:j+filter_size]
filtered_image[i, j] = np.median(window)
return filtered_image
                                                                    #median filter
def apply_median_filter(image_path, filter_sizes):
# Load and normalize image
original_image = np.array(Image.open(image_path).convert('L')) / 255.0
# Add noise to the image
noisy_image = add_noise(original_image)
# Apply median filter with different sizes
filtered_images = [median_filter(noisy_image, size) for size in filter_sizes]
# Plot results
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()
axes[0].imshow(original_image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(noisy_image, cmap='gray')
axes[1].set_title('Noisy Image')
axes[1].axis('off')
for i, (size, filtered) in enumerate(zip(filter_sizes, filtered_images), start=2):
axes[i].imshow(filtered, cmap='gray')
axes[i].set_title(f'Median Filter (size {size}x{size})')
axes[i].axis('off')
plt.tight_layout()
plt.show()

# Example usage
image_path = 'bob.jpg'
filter_sizes = [3, 5, 7, 9]
apply_median_filter(image_path, filter_sizes)
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def add_noise(image, noise_factor=0.1):
"""Add Gaussian noise to the image"""
noisy_image = image + noise_factor * np.random.randn(*image.shape)
return np.clip(noisy_image, 0, 1)

def gaussian_kernel(size, sigma=1):
"""Generate a 2D Gaussian kernel"""
kernel = np.fromfunction(
lambda x, y: (1/(2*np.pi*sigma**2)) *
np.exp(-((x-(size-1)/2)**2 + (y-(size-1)/2)**2)/(2*sigma**2)),
(size, size)
)
return kernel / np.sum(kernel)

def apply_gaussian_filter(image, kernel_size, sigma):
"""Apply Gaussian filter to the image"""
kernel = gaussian_kernel(kernel_size, sigma)
padded_image = np.pad(image, ((kernel_size//2, kernel_size//2),
(kernel_size//2, kernel_size//2)),
mode='reflect')
filtered_image = np.zeros_like(image)
for i in range(image.shape[0]):
for j in range(image.shape[1]):
window = padded_image[i:i+kernel_size, j:j+kernel_size]
filtered_image[i, j] = np.sum(window * kernel)
return filtered_image

def main():
# Load image
image_path = 'sun.jpg'
original_image = np.array(Image.open(image_path).convert('L')) / 255.0

# Add noise to the image
noisy_image = add_noise(original_image, noise_factor=0.1)

# Apply Gaussian filter with different kernel sizes
kernel_sizes = [3, 5, 7, 9]                                                                                  #Gaussian Filter
sigma = 1
filtered_images = [apply_gaussian_filter(noisy_image, size, sigma) for size in kernel_sizes]

# Plot results
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

axes[0].imshow(original_image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(noisy_image, cmap='gray')
axes[1].set_title('Noisy Image')
axes[1].axis('off')

for i, (size, filtered) in enumerate(zip(kernel_sizes, filtered_images), start=2):
axes[i].imshow(filtered, cmap='gray')
axes[i].set_title(f'Gaussian Filter (size {size}x{size}, σ={sigma})')
axes[i].axis('off')

plt.tight_layout()
plt.show()

if __name__ == "__main__":
main()

```
