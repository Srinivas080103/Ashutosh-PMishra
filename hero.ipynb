```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def k_means_clustering(image_path, k=8):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Could not open or find the image.")
        return

    # Original image for reference
    image1 = image.copy()

    # Convert image from BGR to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Create a negative of the image
    negative_image = 255 - image
    image = negative_image

    # Reshape the image into a 2D array of pixels and 3 color values (RGB)
    pixel_values = image.reshape((-1, 3))
    pixel_values = np.float32(pixel_values)

    # Define criteria for k-means clustering
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_COUNT, 100, 0.2)

    # Apply k-means clustering
    _, labels, centers = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

    # Convert cluster centers to integers
    centers = np.uint8(centers)

    # Map the clustered pixels to the image
    clustered_image = centers[labels.flatten()]
    clustered_image = clustered_image.reshape(image.shape)

    # Plot the original and clustered images side by side
    plt.figure(figsize=(12, 6))

    # Original negative image
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title('Negative Image')
    plt.axis('off')

    # Clustered image
    plt.subplot(1, 2, 2)
    plt.imshow(clustered_image)
    plt.title('K-Means Clustered Image')
    plt.axis('off')

    plt.show()

# Call the function with an image path and number of clusters
k_means_clustering('/content/img2.png', k=3)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the input image
image_path = "/content/Picture1.png"
image = cv2.imread(image_path)

if image is None:
    print("Error: Could not open or find the image.")
else:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply mean shift filter to smooth the image while preserving edges
    filtered_image = cv2.pyrMeanShiftFiltering(image, sp=21, sr=51)
    gray_filtered = cv2.cvtColor(filtered_image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding to create a black-and-white image
    _, binary_image = cv2.threshold(gray_filtered, 128, 255, cv2.THRESH_BINARY)

    # Invert the binary image (white text, black background)
    inverted_image = cv2.bitwise_not(binary_image)

    # Apply morphological operations to clean noise and enhance text
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
    cleaned_image = cv2.morphologyEx(inverted_image, cv2.MORPH_OPEN, kernel)
    cleaned_image = cv2.morphologyEx(cleaned_image, cv2.MORPH_CLOSE, kernel)

    # Display results
    plt.figure(figsize=(12, 6))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Mean-shift filtered binary image
    plt.subplot(1, 3, 2)
    plt.imshow(inverted_image, cmap='gray')
    plt.title('Inverted Binary Image')
    plt.axis('off')

    # Cleaned (final) binary image
    plt.subplot(1, 3, 3)
    plt.imshow(cleaned_image, cmap='gray')
    plt.title('Cleaned Image (Final Output)')
    plt.axis('off')

    plt.show()

    # Save the final cleaned image
    output_path = "/content/Output_Image2.png"
    cv2.imwrite(output_path, cleaned_image)
    print(f"Processed image saved to: {output_path}")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the input image (Image 1)
image_path = "/content/Picture1.png"
image = cv2.imread(image_path)

if image is None:
    print("Error: Could not open or find the image.")
else:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply mean shift filter to preserve edges
    shifted_image = cv2.pyrMeanShiftFiltering(image, sp=21, sr=51)

    # Convert the mean-shifted image to grayscale
    gray_shifted = cv2.cvtColor(shifted_image, cv2.COLOR_BGR2GRAY)

    # Threshold the image to create a binary effect
    _, binary_image = cv2.threshold(gray_shifted, 128, 255, cv2.THRESH_BINARY)

    # Invert the binary image to make the text white and background black
    inverted_image = cv2.bitwise_not(binary_image)

    # Display results
    plt.figure(figsize=(12, 6))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Mean-shift filtered image
    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(shifted_image, cv2.COLOR_BGR2RGB))
    plt.title('Mean Shift Filtered')
    plt.axis('off')

    # Final inverted binary image (white text, black background)
    plt.subplot(1, 3, 3)
    plt.imshow(inverted_image, cmap='gray')
    plt.title('Inverted Binary Image')
    plt.axis('off')

    plt.show()

    # Save the output image (white text on black background)
    output_path = "/content/Output_Image2.png"
    cv2.imwrite(output_path, inverted_image)
    print(f"Processed image saved to: {output_path}")
```

Using a Histogram to Convert a Colored Image into a Black and White Image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def convert_to_bw(image_path):
    output_path = "/content/histogram.jpg"
    img = cv2.imread(image_path)

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Histogram equalization
    equalized = cv2.equalizeHist(gray)

    # Otsu's thresholding
    _, bw = cv2.threshold(equalized, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Display images
    plt.figure(figsize=(12, 8))
    plt.subplot(221), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
    plt.subplot(222), plt.imshow(gray, cmap='gray'), plt.title('Grayscale')
    plt.subplot(223), plt.imshow(equalized, cmap='gray'), plt.title('Equalized')
    plt.subplot(224), plt.imshow(bw, cmap='gray'), plt.title('Black & White')
    plt.tight_layout()
    plt.show()

    # Write the black & white image
    cv2.imwrite(output_path, bw)
    print(f"Black and white image saved to {output_path}")

    # Plot histograms
    plt.figure(figsize=(12, 4))
    plt.subplot(131), plt.hist(gray.ravel(), 256, [0, 256]), plt.title('Grayscale Histogram')
    plt.subplot(132), plt.hist(equalized.ravel(), 256, [0, 256]), plt.title('Equalized Histogram')
    plt.subplot(133), plt.hist(bw.ravel(), 256, [0, 256]), plt.title('Black & White Histogram')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    input_image = "/content/img2.png"
    convert_to_bw(input_image)
```

 Apply Non-Linear Filtering Technique to Detect the Edges (Median Filter)

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Import cv2_imshow for image display in Colab

def apply_median_filter_and_detect_edges(image_path):
    # Read the image
    img = cv2.imread(image_path)

    # Display original image using cv2_imshow instead of cv2.imshow
    cv2_imshow(img)
    cv2.waitKey(0)

    # Step 1: Apply Median filter (using a 5x5 kernel)
    median_filtered = cv2.medianBlur(img, 5)  # Kernel size of 5

    # Display the median filtered image
    cv2_imshow(median_filtered)
    cv2.waitKey(0)

    # Step 2: Apply Canny edge detection on the median filtered image
    edges = cv2.Canny(median_filtered, 100, 200)  # Canny edge detection with thresholds 100 and 200

    # Display the edges
    cv2_imshow(edges)
    cv2.waitKey(0)

    # Write the edges to a file
    output_path = "/content/edges_detected_image.jpg"
    cv2.imwrite(output_path, edges)

    # Close all OpenCV windows
    cv2.destroyAllWindows()

if __name__ == "__main__":
    input_image = "/content/img2.png"  # Path to your image
    apply_median_filter_and_detect_edges(input_image)
```
